```python
def crear_driver_chrome(headless=True):
    # for windows
    if name == 'nt':
        PATH = r"chromedriver.exe"

    # for mac and linux(here, os.name is 'posix')
    else:
        PATH = r"./chromedriver"

    options = webdriver.ChromeOptions()
    if headless:
        options.add_argument('--no-sandbox')
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')
        options.add_argument('--disable-dev-shm-usage')
    
    options.add_experimental_option('excludeSwitches', ['enable-logging'])

    options.add_argument('--lang=es')
    options.add_experimental_option('prefs', {'intl.accept_languages': 'es,es_ES'})
    
    driver = webdriver.Chrome(executable_path=PATH, options=options)

    driver.set_window_size(1920, 1080)
    driver.maximize_window()

    return driver
```

```python
def click_element_by_tipo_selector(driver, wait, tipoSelector, textoSelector):
    try:
        elemento = wait.until(EC.presence_of_element_located((tipoSelector, textoSelector)))
        driver.execute_script("arguments[0].click();", elemento)
    except Exception as e:
        pass
```

```python
def click_element_sin_js_by_tipo_selector(driver, wait, tipoSelector, textoSelector):
    try:
        elemento = wait.until(EC.presence_of_element_located((tipoSelector, textoSelector)))
        elemento.click()
    except Exception as e:
        pass
```

```python
def calcular_duracion(contadorinicial):
    contadortimefin = datetime.now()
    segundos = (contadortimefin-contadorinicial).seconds
    dias, restDias = divmod(segundos, 86400)
    hours, restHoras = divmod(restDias, 3600)
    minutes, seconds = divmod(restHoras, 60)
    return 'Duracion: ' + str(dias) + 'd, ' + str(hours) + 'h, ' + str(minutes) + 'm, ' + str(seconds) + 's'
```

```python
def js_function_get_medios_publisuites():
    return """
        let cabeceras = [];
        $('table > thead > tr > th').each(function () {
            cabeceras.push($(this).text().trim() );
        });

        let medios = [];

        if($('table > tbody > tr').length == 1 && $('table > tbody > tr > td').length == 1){
            medios = [];
        } else {
            $('table > tbody > tr').each(function () {
                let medio = new Object();
                medio['tipoMedio'] = tipoMedio;
                    
                $(this).find('td').each(function (index, value) {
            
                    // ['URL', 'DR', 'DA', 'PA', 'CF', 'TF', 'IV', 'Tráfico web/mes', 'RD', 'OBL', 'Indica patrocinado', 'Precio', '']
                    if(cabeceras[index] == 'URL'){
                        medio['url'] = $(this).find('a').text();
                        
                        let pais = $(this).find('img.bandera-tabla').attr('data-original-title');
            
                        if(pais === undefined) {
                            medio['pais'] = null;
                            medio['region'] = null;
                        } else if(pais.indexOf(' - ') !== -1){
                            arrPais = pais.split(' - ');
                            medio['pais'] = arrPais[0];
                            medio['region'] = arrPais[1];
                        } else {
                            medio['pais'] = pais;
                            medio['region'] = null;
                        }
                        
                    } else if(['DR', 'DA', 'PA', 'CF', 'TF'].includes(cabeceras[index])){
                        medio[cabeceras[index]] = parseInt($(this).find('p').text().split('/')[0]);
                    } else if(['IV', 'Tráfico web/mes', 'RD', 'OBL'].includes(cabeceras[index])){
                        medio[cabeceras[index]] = parseFloat($(this).find('p').text());
                    } else if(cabeceras[index] == 'Indica patrocinado'){
                        medio['patrocinado'] = $(this).find('p').text();
                    } else if(cabeceras[index] == 'Precio'){
                        medio['price'] = parseFloat($(this).find('p').text().trim().slice(0, -1).replaceAll('.', ''));
                    }
                    
                });
                
                medios.push(medio);
            });
        }

        return medios;
    """
```

```python
def js_function_get_medios_ids_publisuites():
    return """
        let cabeceras = [];
        $('table > thead > tr > th').each(function () {
            cabeceras.push($(this).text().trim() );
        });

        let medios = [];
        if($('table > tbody > tr').length == 1 && $('table > tbody > tr > td').length == 1){
            medios = [];
        } else {
            $('table > tbody > tr').each(function () {              
                $(this).find('td').each(function (index, value) {
                    if(cabeceras[index] == 'URL'){
                        medios.push($(this).find('a').text());
                    }
                });
            });
        }
        return medios;
    """
```

```python
def get_table_data_publisuites(driver, wait, tipoMedio, mediosOIds='medios'):
    dataFinal = []
    contador = 1
    btnSiguiente = get_element_by_tipo_selector(driver, wait, By.XPATH, '//ul[contains(@class, "pagination")]//a[contains(@class, "page-link-angle") and descendant::i[contains(@class, "fa-angle-right")] and count(*) < 2]')
    while( btnSiguiente != None and btnSiguiente.get_attribute('class').find('disabled') == -1):

        if(mediosOIds == 'medios'):
            dataFinal += driver.execute_script('let tipoMedio="' + tipoMedio + '";' + js_function_get_medios_publisuites())
        else:
            dataFinal += driver.execute_script(js_function_get_medios_ids_publisuites())
        
        print(time.strftime("%d/%m/%y - %H:%M:%S") + ' | ' + tipoMedio + ' | Escaneanda página ' + str(contador) + ' | ' + str(len(dataFinal)) + ' medios acumulados')
        contador += 1

        # Click en Siguiente Página
        driver.execute_script(js_function_scroll_down())
        time.sleep(1)
        btnSiguiente.click()
        esperar_carga_tabla_publisuites(driver)
        time.sleep(2)

        btnSiguiente = get_element_by_tipo_selector(driver, wait, By.XPATH, '//ul[contains(@class, "pagination")]//a[contains(@class, "page-link-angle") and descendant::i[contains(@class, "fa-angle-right")] and count(*) < 2]')
    
    # Recoger Datos de la última página
    if(mediosOIds == 'medios'):
        dataFinal += driver.execute_script('let tipoMedio="' + tipoMedio + '";' + js_function_get_medios_publisuites())
    else:
        dataFinal += driver.execute_script(js_function_get_medios_ids_publisuites())
    print(time.strftime("%d/%m/%y - %H:%M:%S") + ' | ' + tipoMedio + ' | Escaneanda página ' + str(contador) + ' | ' + str(len(dataFinal)) + ' medios acumulados')

    return dataFinal
```

```python
def transformar_arr_en_lista_tuplas_publisuites(arr, url_site):
    lista = []
    contador = 1

    for medio in arr:
        for dato in ['CF', 'DA', 'DR', 'IV', 'OBL', 'PA', 'RD', 'TF', 'Tráfico web/mes', 'pais', 'patrocinado', 'price', 'region', 'url', 'tipoMedio']:
            if not dato in medio:
                medio[dato] = None
        for dato in ['langsselect', 'categoriesselect']:
            if not dato in medio:
                medio[dato] = []

        medio['id'] = contador

        lista.append(
            (
                medio['id'], 
                medio['url'], 
                url_site, 
                medio['CF'], 
                medio['DA'], 
                medio['DR'], 
                medio['IV'], 
                medio['OBL'], 
                medio['PA'], 
                medio['RD'], 
                medio['TF'], 
                medio['Tráfico web/mes'], 
                medio['pais'], 
                medio['patrocinado'], 
                medio['price'], 
                medio['region'], 
                medio['tipoMedio'], 
                '-|-'.join(medio['langsselect']),
                '-|-'.join(medio['categoriesselect'])
            )
        )
        contador += 1

    return lista
```

```python
try:
    # %%
    contadorTimeIniApp = datetime.now()
    escanerConfig = get_config()

    print(time.strftime("%d/%m/%y - %H:%M:%S") + ' | INICIO Scan Publisuites')
    
    driver = crear_driver_chrome(True)
    wait = WebDriverWait(driver, escanerConfig['timeoutSeconds'])

    # %%
    # Abrir Publisuites
    driver.get(escanerConfig['publisuitesUrl'])
    driver.implicitly_wait(escanerConfig['timeoutSeconds'])
    time.sleep(1)


    # %%
    # Escribir usuario
    escribir_text_en_element_by_tipo_selector(driver, wait, By.XPATH, '//input[@name="email"]', escanerConfig['publisuitesUser'])
    time.sleep(1)

    # Escribir contraseña
    escribir_text_en_element_by_tipo_selector(driver, wait, By.XPATH, '//input[@name="password"]', escanerConfig['publisuitesPass'])
    time.sleep(1)

    # Click en Entrar
    click_element_by_tipo_selector(driver, wait, By.XPATH, '//input[@value="Iniciar sesión"]')
    time.sleep(5)


    # %%
    dataFinalArr = {}
    agrupacionesDatosArr = {}

    for tipoMedio in ['Webs & Blogs', 'Periódicos']:
        if tipoMedio == 'Webs & Blogs':
            agrupacionesManualesSelects = ['langsselect', 'categoriesselect']
            textoBtnFiltrar = 'Filtrar webs'
        else:
            agrupacionesManualesSelects = ['langsselect']
            textoBtnFiltrar = 'Filtrar periódicos'
            
        # Click en Webs & Blogs/Periódicos
        driver.execute_script("$('ul.app-menu a:contains(\"" + tipoMedio + "\")')[0].click();")
        time.sleep(5)

        # Click en Mostrar tipo tabla
        click_element_by_tipo_selector(driver, wait, By.XPATH, '//a[contains(@class, "fas fa-grip-lines icon-grid icon-grid m-r-sm   changeview")]')
        esperar_carga_tabla_publisuites(driver)

        # Aplicar filtros generales
        aplicar_filtros_generales_publisuites(driver, wait, tipoMedio)

        # Click en Filtrar webs
        filtrar_webs_publisuites(driver, wait, textoBtnFiltrar)

        # Comprobar estructura tabla que las columnas sigan en el mismo orden y las mismas columnas
        estrcturaNueva = driver.execute_script(js_function_get_cabeceras_tabla_publisuites())

        if(not comprobar_estructura_tabla_publisuites(estrcturaNueva)):
            raise Exception("Han cambiado las columnas, o el orden o falta/se ha añadido otra columna en la tabla de publisuites")
        
        print(time.strftime("%d/%m/%y - %H:%M:%S") + ' | Escaneando todos los ' + tipoMedio)
        dataFinalArr[tipoMedio] = get_table_data_publisuites(driver, wait, tipoMedio, 'medios')
        time.sleep(4)

        agrupacionesDatosArr[tipoMedio] = {}

        for agrupacion in agrupacionesManualesSelects:

            # Recoger todos los idiomas/categorias disponibles
            agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion] = driver.execute_script(js_function_get_select_options('#' + agrupacion + ' option', 'text'))
            
            if agrupacion == 'langsselect':
                auxList = []
                for elem in agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion]:
                    if elem in escanerConfig['publisuitesLanguageIncluir']: auxList.append(elem)
                agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion] = auxList.copy()
            elif agrupacion == 'categoriesselect':
                for elem in escanerConfig['publisuitesCategoryFilterExcluir']:
                    if elem in agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion]: agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion].remove(elem)

            # Cargar cada idioma/categoria y recoger las ids de los medios que aparecen en la tabla por cada filtro de cada idioma
            agrupacionesDatosArr[tipoMedio]['tipo_ids_' + agrupacion] = {}
            for agrupacionTipo in agrupacionesDatosArr[tipoMedio]['tipo_' + agrupacion]:

                # Aplicar filtros generales
                aplicar_filtros_generales_publisuites(driver, wait, tipoMedio)

                # Desplegar menú Tipos de enlaces
                click_element_by_tipo_selector(driver, wait, By.XPATH, '//button[@data-id="' + agrupacion + '"]')
                time.sleep(1)

                # Click en Dato desplegado
                click_element_by_tipo_selector(driver, wait, By.XPATH, '//li//a//span[contains(text(),"' + agrupacionTipo + '")]')
                time.sleep(1)

                # Quitar menú Tipos de enlaces
                click_element_by_tipo_selector(driver, wait, By.XPATH, '//button[@data-id="' + agrupacion + '"]')
                time.sleep(1)

                # Click en Filtrar webs
                filtrar_webs_publisuites(driver, wait, textoBtnFiltrar)
                
                driver.execute_script(js_function_scroll_down())
                time.sleep(2)
                
                print(time.strftime("%d/%m/%y - %H:%M:%S") + ' | ' + tipoMedio + ' | Escaneando ' + agrupacion + ' -> ' + agrupacionTipo)
                agrupacionesDatosArr[tipoMedio]['tipo_ids_' + agrupacion][agrupacionTipo] = get_table_data_publisuites(driver, wait, tipoMedio, 'ids')
                time.sleep(2)

            
            # Recorrer cada agrupación una vez y crear una lista auxiliar con los ids de cada medio con el grupo/grupos al que pertenece (opción más eficiente)
            # Que luego por cada id recorrer cada agrupación (Idioma/Categoría) para ver si está dentro
            agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion] = {}
            for datoAgrupacion in agrupacionesDatosArr[tipoMedio]['tipo_ids_' + agrupacion]:
                for idMedio in agrupacionesDatosArr[tipoMedio]['tipo_ids_' + agrupacion][datoAgrupacion]:
                    if not idMedio in agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion]:
                        agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion][idMedio] = []
                    agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion][idMedio].append(datoAgrupacion)

        

    # %%
    dataFinal = []
    for unDataFinal in dataFinalArr:
        dataFinal.extend(dataFinalArr[unDataFinal])

    # %%
    for idData, data in enumerate(dataFinal):
        tipoMedio = dataFinal[idData]['tipoMedio']
        if tipoMedio == 'Webs & Blogs':
            agrupacionesManualesSelects = ['langsselect', 'categoriesselect']
        else:
            agrupacionesManualesSelects = ['langsselect']
        
        agrupacionesManuales = []
        agrupacionesManuales.extend(agrupacionesManualesSelects)

        for agrupacion in agrupacionesManuales:
            if dataFinal[idData]['url'] in agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion]:
                if not agrupacion in dataFinal[idData]:
                    dataFinal[idData][agrupacion] = []
                dataFinal[idData][agrupacion].extend(agrupacionesDatosArr[tipoMedio]['ids_tipo_' + agrupacion][dataFinal[idData]['url']])

    # %%
    dataFinalTuplas = transformar_arr_en_lista_tuplas_publisuites(dataFinal, escanerConfig['publisuitesUrl'])

    # %%
    # Conexion MySQL
    mydb = get_mysql_db(escanerConfig)
    mycursor = mydb.cursor()

    # Crear Tabla si no existe, si existe eliminar todas las filas
    print(time.strftime("%d/%m/%y - %H:%M:%S") + ' Crear Tabla ' + escanerConfig['dbTablePublisuites'] + ' si no existe, si existe eliminar todas las filas')
    mycursor.execute('SELECT count(*) FROM information_schema.TABLES WHERE (TABLE_SCHEMA = \'' + escanerConfig['dbName'] + '\') AND (TABLE_NAME = \'' + escanerConfig['dbTablePublisuites'] + '\')')
    myresult = mycursor.fetchone()
    if myresult[0] == 0:
        mycursor.execute(get_sql_crear_tabla_medio_publisuites(escanerConfig['dbTablePublisuites']))
    else:
        mycursor.execute('TRUNCATE TABLE ' + escanerConfig['dbTablePublisuites'])

    # Insertar
    mycursor.executemany(get_sql_insert_tabla_medio_publisuites(escanerConfig['dbTablePublisuites']), dataFinalTuplas)
    mydb.commit()
    print(mycursor.rowcount, ' medios introducidos correctamente en tabla ' + escanerConfig['dbTablePublisuites'] + '.')
    
    mycursor.close()
    mydb.close()


    # %%
    driver.close()
    driver.quit()


# %%
except Exception as e:
    print("Something went wrong:\n" + str(e))
  
```
