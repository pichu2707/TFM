```python
def crear_driver_chrome(headless=True):
    # for windows
    if name == 'nt':
        PATH = r"chromedriver.exe"

    # for mac and linux(here, os.name is 'posix')
    else:
        PATH = r"./chromedriver"

    options = webdriver.ChromeOptions()
    if headless:
        options.add_argument('--no-sandbox')
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')
        options.add_argument('--disable-dev-shm-usage')
    
    options.add_experimental_option('excludeSwitches', ['enable-logging'])

    options.add_argument('--lang=es')
    options.add_experimental_option('prefs', {'intl.accept_languages': 'es,es_ES'})
    
    driver = webdriver.Chrome(executable_path=PATH, options=options)

    driver.set_window_size(1920, 1080)
    driver.maximize_window()

    return driver
```

```python
js_function_get_medios_publisuites = driver.execute_script("""
        let cabeceras = [];
        $('table > thead > tr > th').each(function () {
            cabeceras.push($(this).text().trim() );
        });

        let medios = [];

        if($('table > tbody > tr').length == 1 && $('table > tbody > tr > td').length == 1){
            medios = [];
        } else {
            $('table > tbody > tr').each(function () {
                let medio = new Object();
                let tipoMedio = 'Webs & Blogs'
                medio['tipoMedio'] = tipoMedio;
                    
                $(this).find('td').each(function (index, value) {
            
                    // ['URL', 'DR', 'DA', 'PA', 'CF', 'TF', 'IV', 'Tráfico web/mes', 'RD', 'OBL', 'Indica patrocinado', 'Precio', '']
                    if(cabeceras[index] == 'URL'){
                        medio['url'] = $(this).find('a').text();
                        
                        let pais = $(this).find('img.bandera-tabla').attr('data-original-title');
            
                        if(pais === undefined) {
                            medio['pais'] = null;
                            medio['region'] = null;
                        } else if(pais.indexOf(' - ') !== -1){
                            arrPais = pais.split(' - ');
                            medio['pais'] = arrPais[0];
                            medio['region'] = arrPais[1];
                        } else {
                            medio['pais'] = pais;
                            medio['region'] = null;
                        }
                        
                    } else if(['DR', 'DA', 'PA', 'CF', 'TF'].includes(cabeceras[index])){
                        medio[cabeceras[index]] = parseInt($(this).find('p').text().split('/')[0]);
                    } else if(['IV', 'Tráfico web/mes', 'RD', 'OBL'].includes(cabeceras[index])){
                        medio[cabeceras[index]] = parseFloat($(this).find('p').text());
                    } else if(cabeceras[index] == 'Indica patrocinado'){
                        medio['patrocinado'] = $(this).find('p').text();
                    } else if(cabeceras[index] == 'Precio'){
                        medio['price'] = parseFloat($(this).find('p').text().trim().slice(0, -1).replaceAll('.', ''));
                    }
                    
                });
                
                medios.push(medio);
            });
        }

        return medios;
    """)

print(js_function_get_medios_publisuites)
```
